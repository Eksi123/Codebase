```python
# 统计市长电话工单文本长度（以词为单位）
import numpy as np
import jieba
import matplotlib.pyplot as plt

doc=open('./data/data2005-2020.txt','r',encoding='utf-8')
Count = []
text = doc.readline().rstrip()

while True:
    text = doc.readline().rstrip()
    if not text:
        break
    text = text.split("\t")
    text = text[len(text)-1]
    word_list = jieba.cut(text)
    word_list = ("/".join(word_list)).split('/')
    Count.append(len(word_list))

Count = np.array(Count)
print(Count.mean(),Count.std())
plt.hist(Count,bins=50, range=[0,300],density=True)

"""
mean:54.53664826025758
std:39.81126750872251
"""
```

```python
# 提取关键词（服务态度与质量、工作效率）

from jieba import analyse

text1 = open("data/服务态度与质量（二级行业）/2020.txt","r",encoding='utf-8').read()
text2 = open('data/工作效率（二级行业）/2020.txt','r',encoding='utf-8').read()

keywords1 = analyse.extract_tags(text1, topK=50, withWeight=True, allowPOS=('n', 'v','vn'))
keywords2 = analyse.extract_tags(text2, topK=50, withWeight=True, allowPOS=('n', 'v','vn'))

print(keywords1)
print(keywords2)
```
